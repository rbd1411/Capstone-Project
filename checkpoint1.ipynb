```python
import pandas as pd
from scipy.stats import zscore

# Replace 'data_summer_course.csv' with your actual file path
dataset_path = 'data_summer_course.csv'

# Load the dataset into a DataFrame
data_df = pd.read_csv(dataset_path)

# Handling Missing Values
numerical_columns = data_df.select_dtypes(include=['float64']).columns
data_df[numerical_columns] = data_df[numerical_columns].fillna(data_df[numerical_columns].mean())

# Handling Outliers
z_scores = zscore(data_df.select_dtypes(include=['float64']))  # Calculate z-scores for numerical columns
data_df = data_df[(z_scores < 3).all(axis=1)]  # Remove rows with outlier z-scores

# Handling Inconsistent Data (example: correcting device types)
data_df['DEVICETYPE'] = data_df['DEVICETYPE'].str.replace('Smarphone', 'Smartphone', case=False)

# Display cleaned DataFrame information
print("Cleaned DataFrame:")
print(data_df.head())
print()

# Display summary statistics of cleaned numerical columns
print("Summary statistics of cleaned numerical columns:")
print(data_df.describe())
print()

# Display the dimensions of the cleaned DataFrame
print("Dimensions of the cleaned DataFrame:")
print(data_df.shape)
```
