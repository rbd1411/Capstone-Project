## c. Data cleaning: Identify and handle missing values, outliers, and inconsistent data. Implement appropriate techniques to clean the data, such as dropping or imputing missing values, removing outliers, and addressing inconsistent entries.

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('data_summer_course.csv')

# Drop columns with many missing values
columns_to_drop = ['last_app_used', 'CNTCHURND1_30', 'HANDSETCHANGESFLAGD1_30', 'DEVICENETWORK', 'DEVICEMODELC', 'DEVICEDUALSIMFLAG']
df_cleaned = df.drop(columns=columns_to_drop)

# Forward-fill missing date values
date_columns = ['LASTVCEUSGDATE', 'LASTRECHRGDATE', 'FIRSTCALLDATE', 'LASTREVGENEVTDATE']
df_cleaned[date_columns] = df_cleaned[date_columns].fillna(method='ffill')

# Convert date columns to datetime format
df_cleaned[date_columns] = df_cleaned[date_columns].apply(pd.to_datetime, errors='coerce')

# Drop rows with missing values in specific columns
df_cleaned = df_cleaned.dropna(subset=['SUBSRIBERLASTBALANCE', 'CUSTSEGMENT', 'SMARTPHONEFLAG'])

# Handle outliers in MAINACTBAL column (replace with NaN)
mainactbal_upper_bound = df_cleaned['MAINACTBAL'].quantile(0.99)
df_cleaned.loc[df_cleaned['MAINACTBAL'] > mainactbal_upper_bound, 'MAINACTBAL'] = np.nan

# Perform categorical data cleaning (convert to uppercase)
categorical_columns = ['DEVICETYPE', 'ISDEVICE3GENABLED', 'VASSUBSCRIBERFLAG', 'ISDEVICEDATAENABLED', 'CUSTSEGMENT', 'SMARTPHONEFLAG']
df_cleaned[categorical_columns] = df_cleaned[categorical_columns].apply(lambda x: x.str.upper())

# Handle missing values in remaining columns (impute with median or mode)
df_cleaned['AGEONNETWORK'] = df_cleaned['AGEONNETWORK'].fillna(df_cleaned['AGEONNETWORK'].median())
df_cleaned['ARPUD1_30'] = df_cleaned['ARPUD1_30'].fillna(df_cleaned['ARPUD1_30'].median())

# Confirm the cleaned data
print(df_cleaned.info())
print(df_cleaned.describe())
```

## d. Data transformation: Apply necessary transformations to the data to make it suitable for analysis. This may include feature scaling, encoding categorical variables, creating derived variables, or aggregating data as required.

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pandas as pd

# Load your cleaned dataset
# df_cleaned = ...

# 1. Feature Scaling (using StandardScaler)
scaler = StandardScaler()
scaled_columns = ["MAINACTBAL", "AGEONNETWORK", "SUBSRIBERLASTBALANCE", "ARPUD1_30"]
df_cleaned[scaled_columns] = scaler.fit_transform(df_cleaned[scaled_columns])

# 2. Encoding Categorical Variables (using Label Encoding)
label_encoder = LabelEncoder()
categorical_columns = ["DEVICETYPE", "ISDEVICE3GENABLED", "VASSUBSCRIBERFLAG", "ISDEVICEDATAENABLED", "CUSTSEGMENT", "SMARTPHONEFLAG"]
df_cleaned[categorical_columns] = df_cleaned[categorical_columns].apply(label_encoder.fit_transform)

# 3. Handling Missing Values
# You can choose to impute missing values with the mean or median of the respective column
imputed_columns = ["MAINACTBAL", "SUMVCEREVCD1_30", "SUMDATAREVCD1_30", "SUMDATAUSGCD1_30", "SUMVOICEUSGCD1_30", "SUMDATAUSG4GCD1_15", "SUMDATAUSG4GCD15_30"]
df_cleaned[imputed_columns] = df_cleaned[imputed_columns].fillna(df_cleaned[imputed_columns].median())
```

## e. Data visualization: Utilize Python's data visualization libraries, such as matplotlib or seaborn, to create informative visualizations. Generate various types of plots, such as histograms, bar charts, scatter plots, or box plots, to understand the distribution, relationships, and patterns within the dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Histogram of MAINACTBAL
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned["MAINACTBAL"], bins=50, kde=True)
plt.title("Distribution of Main Account Balance")
plt.xlabel("Main Account Balance")
plt.ylabel("Frequency")
plt.show()

# Box Plot of MAINACTBAL by CUSTSEGMENT
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_cleaned, x="CUSTSEGMENT", y="MAINACTBAL")
plt.title("Box Plot of Main Account Balance by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Main Account Balance")
plt.xticks(rotation=45)
plt.show()

# Scatter Plot of AGEONNETWORK vs. MAINACTBAL colored by SMARTPHONEFLAG
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned, x="AGEONNETWORK", y="MAINACTBAL", hue="SMARTPHONEFLAG", alpha=0.5)
plt.title("Scatter Plot of Age on Network vs. Main Account Balance")
plt.xlabel("Age on Network")
plt.ylabel("Main Account Balance")
plt.legend()
plt.show()

# Count Plot of CUSTSEGMENT
plt.figure(figsize=(10, 6))
sns.countplot(data=df_cleaned, x="CUSTSEGMENT")
plt.title("Count Plot of Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# Correlation Heatmap
plt.figure(figsize=(12, 8))
corr_matrix = df_cleaned.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Time Series Plot of MAINACTBAL over time
df_time_series = df_cleaned.set_index("LASTVCEUSGDATE")
plt.figure(figsize=(12, 8))
sns.lineplot(data=df_time_series, x=df_time_series.index, y="MAINACTBAL")
plt.title("Main Account Balance Over Time")
plt.xlabel("Date")
plt.ylabel("Main Account Balance")
plt.xticks(rotation=45)
plt.show()
```

## g. Exploratory Data Analysis (EDA): Conduct EDA techniques to uncover patterns, trends, and relationships within the data. Use visualizations to understand the factors influencing customer behavior, such as call duration, plan usage, or campaign response rates.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Analyze call duration

# Scatter Plot of AGEONNETWORK vs. SUMVCEREVCD1_30
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned, x="AGEONNETWORK", y="SUMVCEREVCD1_30", alpha=0.5)
plt.title("Age vs. Voice Call Revenue")
plt.xlabel("Age on Network")
plt.ylabel("Voice Call Revenue")
plt.show()

# Analyze plan usage

# Pair Plot of selected usage variables
usage_vars = ["SUMVCEREVCD1_30", "SUMDATAREVCD1_30", "SUMDATAUSGCD1_30", "SUMVOICEUSGCD1_30"]
sns.pairplot(df_cleaned[usage_vars])
plt.suptitle("Pair Plot of Usage Variables", y=1.02)
plt.show()

# Analyze campaign response rates

# Bar Plot of VASSUBSCRIBERFLAG by CUSTSEGMENT
plt.figure(figsize=(10, 6))
sns.barplot(data=df_cleaned, x="CUSTSEGMENT", y="SUMVCEREVCD1_30", hue="VASSUBSCRIBERFLAG", ci=None)
plt.title("Voice Call Revenue by Customer Segment and VAS Subscriber Flag")
plt.xlabel("Customer Segment")
plt.ylabel("Voice Call Revenue")
plt.xticks(rotation=45)
plt.legend(title="VAS Subscriber Flag")
plt.show()

# Heatmap of correlation between usage variables
usage_corr = df_cleaned[usage_vars].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(usage_corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Usage Variables")
plt.show()
```
